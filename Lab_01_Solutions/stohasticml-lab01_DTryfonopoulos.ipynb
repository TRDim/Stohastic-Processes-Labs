### C)Θα απεικονίζει τα σημεία που προκύπτουν ως ζεύγη της ανεξάρτητης και εξαρτημένης μεταβλητής που απομονώθηκαν στο πρώτο ερώτημα. Τι παρατηρείτε σε σχέση με τα συμπεράσματα που βγάλατε στο δεύτερο ερώτημα;

Θα υπολογίσουμε το Covariance απο 2 γκρουπ αριθμων , το οποιο περιγραφει πως αυτοι οι 
αριθμοι αλλαζουν ταυτοχρονα. Το covariance ειναι μια γενίκευση του correlation. To 
correlation περιγραφει την σχεση μεταξυ 2 γκρουπ αριθμων, οπου το covariance μπορει να
χρησιμοιποιηθει και για την περιγραφη 2 αλλα και περισσοτερων γκρουπ αριθμων. 

# covariance = sum((x(i)- mean(x)) * (y(i) - mean(y))) / (N-1)

```python
# Υπολογισμός Covariance & Variance

def covariance (x, mean_x, y, mean_y):
    covar=0.0
    for i in range (len(x)):
        covar += ((x[i] - mean_x) * (y[i] - mean_y))
    covar = covar/ (len(x)-1)
    return covar 

def variance(values, mean_val):
    var = 0
    for x in values:
        var+= (x- mean_val)**2
    var = var/(len(values)-1)
    return var

def stadDev(values, mean_val):
    sumM=0
    for i in range(len(values)):
        sumM += (values[i] - mean_val)**2
    
    stadDev= np.sqrt(sumM / len(values))
    return stadDev         
```

```python
def PearCorCo (x, mean_x, y, mean_y):
    PearCorCo = covariance (x, mean_x, y, mean_y) / (stadDev(x, mean_x) *stadDev(y, mean_y))
    return PearCorCo
```

### E) Θα κανονικοποιεί τα δεδομένα που παρέχονται με τη μέθοδο min-max normalization. Περισσότερες πληροφορίες για τη μέθοδο αυτή μπορούν να βρεθούν

```python
# Split the Data in 80%-Train & 20%-Test
len1 = len(dataset1.values)
len2 = len(dataset2.values)

split1 = int(len1 * 0.8)
split2 = int(len2 * 0.8)

# Dataset1
a1_Train, b1_Train, a1_Test, b1_Test = a1[:split1], b1[:split1],a1[split1:], b1[split1:] 


# Dataset2
a2_Train, b2_Train, a2_Test, b2_Test = a2[:split2], b2[:split2],a2[split2:], b2[split2:] 
```

### Min-Max Scaler

X_Norm = (X - X.min(axis=0)) / (X.max(axis=0) - X.min(axis=0))


```python
def scaleData(data):
    data_N = 0* (data)
    minD= min(data)
    maxD = max(data)

    for i in range(0, len(data)):
        data_N[i,] = (data[i] - minD) / (maxD- minD)
         
    return data_N
        
```

```python
# Dataset1 Scaled
a1_Train_N=  scaleData(a1_Train)
b1_Train_N=  scaleData(b1_Train)
a1_Test_N =  scaleData(a1_Test)
b1_Test_N =  scaleData(b1_Test)

# Dataset2 Scaled
a2_Train_N=  scaleData(a2_Train)
b2_Train_N=  scaleData(b2_Train)
a2_Test_N =  scaleData(a2_Test)
b2_Test_N =  scaleData(b2_Test)
```

### F) Θα υπολογίζει τις παραμέτρους του μοντέλου linear regression, χρησιμοποιώντας τη μέθοδο Stochastic Gradient Descent (SGD). Για το σκοπό αυτό, ο χρήστης θα πρέπει να ορίζει το ρυθμό μάθησης (learning rate) και το συνολικό αριθμό των εποχών (epochs) εκπαίδευσης. ### G) Θα υπολογίζει το μέσο τετραγωνικό σφάλμα (mean squared error) των προβλέψεων πάνω στο test set. Περισσότερες πληροφορίες μπορούν να βρεθούν 

```python
def MyCustomSGD(train_data,learning_rate,n_iter,k,divideby, dataset1=True):
    
    # Initially we will keep our W and B as 0 as per the Training Data
    w=np.zeros(shape=(1,train_data.shape[1]-1))
    b=0
    
    cur_iter=1
    while(cur_iter<=n_iter): 

        # We will create a small training data set of size K
        temp=train_data.sample(k)
        
        # We create our X and Y from the above temp dataset
        if dataset1:
            y=np.array(temp['likes'])
            x=np.array(temp.drop('likes',axis=1))
        else:
            y=np.array(temp['y'])
            x=np.array(temp.drop('y',axis=1))
        
        # We keep our initial gradients as 0
        w_gradient=np.zeros(shape=(1,train_data.shape[1]-1))
        b_gradient=0
        
        for i in range(k): # Calculating gradients for point in our K sized dataset
            prediction=np.dot(w,x[i])+b
            w_gradient=w_gradient+(-2)*x[i]*(y[i]-(prediction))
            b_gradient=b_gradient+(-2)*(y[i]-(prediction))
        
        #Updating the weights(W) and Bias(b) with the above calculated Gradients
        w=w-learning_rate*(w_gradient/k)
        b=b-learning_rate*(b_gradient/k)
        
        # Incrementing the iteration value
        cur_iter=cur_iter+1
        
        #Dividing the learning rate by the specified value
        learning_rate=learning_rate/divideby
        
    return w,b #Returning the weights and Bias
```

```python
learning_rate = 0.2
n_iter=100
k=100
divideby=10
w1, b1 = MyCustomSGD(TrainData1i,learning_rate,n_iter,k,divideby)
print(w1,b1)
```

```python
learning_rate = 0.2
n_iter=100
k=100
divideby=10
w2, b2 = MyCustomSGD(TrainData2i,learning_rate,n_iter,k,divideby, dataset1=False)
print(w2, b2)
```

```python
def predict(x,w,b):
    y_pred=[]
    for i in range(len(x)):
        y=np.asscalar(np.dot(w,x[i])+b)
        #y=(np.dot(w,x[i])+b).item()
        y_pred.append(y)
    return np.array(y_pred)
```

```python
def mean_squared_error(test, predicted):
    sumMSE =0
    for i in range(len(test)):
        sumMSE += (test[i]-predicted[i])**2
    MSE= sumMSE/len(test)
    return MSE
    
```

```python
train_data = TrainData1i
x_test = np.array(TestData1i['views'])
y_test = np.array(TestData1i['likes'])
w,b=MyCustomSGD(train_data,learning_rate=1,n_iter=100,divideby=2,k=10)
y_pred_customsgd=predict(x_test,w,b)

plt.scatter(y_test,y_pred_customsgd)
plt.grid()
plt.xlabel('Actual y')
plt.ylabel('Predicted y')
plt.title('Scatter plot from actual y and predicted y')
plt.show()
print('Mean Squared Error :',mean_squared_error(y_test, y_pred_customsgd))
```
